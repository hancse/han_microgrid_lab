```python
def markVoltNaN(inVolt):
    if (inVolt < -300) or (inVolt > 300):
        return np.NaN
    return inVolt

def markAmpreNaN(inAmp):
    if (inAmp < -30) or (inAmp > 30):
        return np.NaN
    return inAmp

def setIndextoTime(data_in):
    data_out = data_in
    
    data_out['V1'] = data_in['V1'].map(markVoltNaN)
    data_out['V2'] = data_in['V2'].map(markVoltNaN)
    data_out['V3'] = data_in['V3'].map(markVoltNaN)
    
    data_out['I1'] = data_in['I1'].map(markAmpreNaN)
    data_out['I2'] = data_in['I2'].map(markAmpreNaN)
    data_out['I3'] = data_in['I3'].map(markAmpreNaN)
    
    data_out = data_out.fillna(method='ffill')
    
    data_out.loc[:,'Time'] = pd.to_datetime(data_out.loc[:,'Time'],unit='s')
    return data_out.set_index(['Time'])
```
