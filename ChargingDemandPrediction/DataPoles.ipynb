```python
def markVoltNaN(inVolt):
    if (inVolt < -300) or (inVolt > 300):
        return np.NaN
    return inVolt

def markAmpreNaN(inAmp):
    if (inAmp < -30) or (inAmp > 30):
        return np.NaN
    return inAmp

def setIndextoTime(data_in):
    data_out = data_in
    
    data_out['V1'] = data_in['V1'].map(markVoltNaN)
    data_out['V2'] = data_in['V2'].map(markVoltNaN)
    data_out['V3'] = data_in['V3'].map(markVoltNaN)
    
    data_out['I1'] = data_in['I1'].map(markAmpreNaN)
    data_out['I2'] = data_in['I2'].map(markAmpreNaN)
    data_out['I3'] = data_in['I3'].map(markAmpreNaN)
    
    data_out = data_out.fillna(method='ffill')
    
    data_out.loc[:,'Time'] = pd.to_datetime(data_out.loc[:,'Time'],unit='s')
    data_out.set_index(['Time'], inplace=True)
    data_out.sort_index(inplace=True)
    data_out.loc[:,'Time_diff'] = (data_out.index.to_series().diff()).dt.total_seconds()
    return data_out
```

```python
''' If the time diff is larger than 600s (10 minutes), then data are assumed loss and E is 0 '''

def removeLargeDiff(diff_in):
    if (diff_in > 600):
        return 0
    return diff_in

def calculateEnergy(data_in):
    data_out = data_in.copy()
    data_out['Power'] = data_out['I1']*data_out['V1'] + data_out['I2']*data_out['V2'] + data_out['I3']*data_out['V3']
    data_out.loc[:,'Time_diff'] = data_out.Time_diff.shift(-1)
    data_out = data_out.iloc[:-1,:]
    data_out['Time_diff'] = data_out.Time_diff.map(removeLargeDiff)
    data_out['Energy_kWh'] = data_out['Power'] * data_out['Time_diff'] / 3600000
    data_out = data_out[['Energy_kWh']].groupby(pd.Grouper(freq='1H')).sum()
    return data_out
    
```
